Take the derivative of the product of two symbolic functions to get an expression for product rule.

```python
## Your code here

from functions import Symbolic, Product

f = Symbolic("f")
g = Symbolic("g")
left = Symbolic("(f*g)")
p = Product(f, g)
print("right hand side of equation: " + p.derivative()("x"))
print(left.derivative()("x") + " = " + p.derivative()("x"))
```

2. Derive a rule for the derivative of $f = g/h$ using symbolic functions.  Does this reduce to quotient rule?

```python
## Your code here
left = Symbolic("f")
right = Symbolic("g") / Symbolic("h")
print("left side: " + left.derivative()("x"))
print("right side: " + right.derivative()("x"))
print(left.derivative()("x") + " = " + right.derivative()("x"))
```

Yes, this reduces to quotient rule.
